```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
pyd = pd.read_csv("div_test.py.out", sep=" ", names=["nbits", "dbits", "time"])
jnd = pd.read_csv("div_test.jn.out", sep=" ", names=["nbits", "dbits", "time"])
df = pd.merge(pyd, jnd, how="left",
              left_on=["nbits", "dbits"],
              right_on=["nbits", "dbits"],
              suffixes=["_py","_jn"],
             )
df
```

```python
df2 = pd.DataFrame({'python': df.time_py, 'libbf': df.time_jn,
                   'index': df.nbits.astype(str) + "/" + df.dbits.astype(str)},
                   )
df2.plot(kind="bar", x='index', logy=True)
```
